- Langkah pembersihan dilakukan untuk memastikan data yang digunakan bersih dan konsisten
- Menghapus nilai hilang di kolom penting (country, agent, company) untuk menjaga relevansi analisis.
- Mengimputasi nilai hilang menggunakan rata-rata (numerik) dan modus (kategorikal) untuk menghindari kehilangan data yang signifikan.
- Menghapus duplikasi untuk memastikan setiap data unik.


```python
#Langkah-Langkah Pembersihan Data
cleaned_dataframes = {}

for file_name, df in dataframes.items():
    # a. Hapus nilai hilang pada kolom tertentu
    for col in ['country', 'agent', 'company']:
        if col in df.columns:
            df = df[df[col].notnull()]

    # b. Imputasi nilai hilang
    for col in df.columns:
        if df[col].dtype in ['float64', 'int64']:
            df[col].fillna(df[col].mean(), inplace=True)
        elif df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)

    # c. Konversi kolom tanggal (jika ada)
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')

    # d. Hapus duplikasi
    df.drop_duplicates(inplace=True)

    # Simpan DataFrame yang telah dibersihkan
    cleaned_dataframes[file_name] = df
```

```python
for file_name, df in cleaned_dataframes.items():
    print(f"\nSummary of {file_name}:")
    print(f"Jumlah baris: {df.shape[0]}")
    print(f"Jumlah kolom: {df.shape[1]}")
    print("Kolom yang tersedia:", df.columns.tolist())
    print("Nilai hilang per kolom:", df.isnull().sum().to_dict())
```

```python
#Tampilkan Kumpulan Data Akhir (Ringkas)
for file_name, df in cleaned_dataframes.items():
    print(f"\nDisplaying cleaned data from {file_name} as a DataFrame:")
    display(df.head())
```

#Eksplorasi dan analisa data

```python
for file_name, df in cleaned_dataframes.items():
    print(f"\nAnalyzing {file_name}:")

    if 'adr' in df.columns and 'assigned_room_type' in df.columns:
        # Menambahkan kolom baru: rasio ADR terhadap rata-rata ADR untuk setiap tipe kamar
        df['adr_to_avg_ratio'] = df['adr'] / df.groupby('assigned_room_type')['adr'].transform('mean')

        # Menampilkan rasio rata-rata untuk setiap tipe kamar
        adr_ratio_mean = df.groupby('assigned_room_type')['adr_to_avg_ratio'].mean()
        print("Average ADR to Average Ratio by Assigned Room Type:")
        print(adr_ratio_mean)
```

```python
for file_name, df in cleaned_dataframes.items():
    print(f"\nAnalyzing ADR and Assigned Room Type in {file_name}:")
    if 'adr' in df.columns and 'assigned_room_type' in df.columns:

        #Rata-rata ADR berdasarkan Assigned Room Type
        room_adr_mean = df.groupby('assigned_room_type')['adr'].mean().sort_values(ascending=False)
        print("\nAverage ADR by Assigned Room Type:")
        print(room_adr_mean)

        # Plot rata-rata ADR
        plt.figure(figsize=(10, 6))
        sns.barplot(x=room_adr_mean.index, y=room_adr_mean.values, palette="coolwarm")
        plt.title(f"Average ADR by Assigned Room Type in {file_name}")
        plt.xlabel("Assigned Room Type")
        plt.ylabel("Average ADR")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
    if 'country' in df.columns:
        # Menemukan 10 negara dengan kontribusi ADR tertinggi
        country_adr = df.groupby('country')['adr'].mean().sort_values(ascending=False).head(10)
        print("Top 10 Countries by Average ADR:")
        print(country_adr)

        # Plot top 10 countries by ADR
        plt.figure(figsize=(12, 6))
        sns.barplot(x=country_adr.index, y=country_adr.values, palette="viridis")
        plt.title(f"Top 10 Countries by Average ADR in {file_name}")
        plt.xlabel("Country")
        plt.ylabel("Average ADR")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
#Distribusi ADR untuk setiap Assigned Room Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='assigned_room_type', y='adr', data=df, palette="viridis")
plt.title(f"ADR Distribution by Assigned Room Type in {file_name}")
plt.xlabel("Assigned Room Type")
plt.ylabel("ADR")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Korelasi ADR dan Assigned Room Type
df['room_type_numeric'] = df['assigned_room_type'].astype('category').cat.codes
correlation = df[['adr', 'room_type_numeric']].corr().iloc[0, 1]
print(f"\nCorrelation between ADR and Assigned Room Type in {file_name}: {correlation:.2f}")
```
